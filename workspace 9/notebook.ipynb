## Getting Started
I first uploaded the three files I will be working with as DataFrames. I wasn't sure how many sheets were in `data/airbnb_room_type.xlsx`, so I checked first before converting it to a DataFrame. 

I edited the 'price' column in `airbnb_price` by removing the ' dollars' string and changing the dtype to 'float' to make that column easier to work with in my analysis. 

I then checked all the unique entries in the 'room_type' column of `airbnb_room_type` to determine the number of different room types. Upon checking, I found there were various types of capitalization for the same room types, causing them to appear as different room types. To consolidate, I converted all entries to title case. leaving me with three unique room types. 

Lastly, I converted the 'last_review' column of `last_review` to a datetime object to make it easier to work with in my analysis. I also checked to make sure `pd.to_datetime` was able to convert all the dates. 

```python
airbnb_price = pd.read_csv('data/airbnb_price.csv')
airbnb_price.head()
```

```python
airbnb_price['price'] = airbnb_price['price'].str.strip(' dollars')
airbnb_price['price'] = airbnb_price['price'].astype('float')
airbnb_price.head()
```

```python
airbnb_room_type_excel = pd.ExcelFile('data/airbnb_room_type.xlsx')
print(airbnb_room_type_excel.sheet_names)
airbnb_room_type = airbnb_room_type_excel.parse('airbnb_room_type')
airbnb_room_type.head()
```

```python
print(airbnb_room_type['room_type'].unique())
airbnb_room_type['room_type'] = airbnb_room_type['room_type'].str.title()
print(airbnb_room_type['room_type'].unique())
```

```python
airbnb_last_review = pd.read_csv('data/airbnb_last_review.tsv', sep = '\t')
airbnb_last_review.head()
```

```python
airbnb_last_review['last_review'] = pd.to_datetime(airbnb_last_review['last_review'], errors='coerce')
print(airbnb_last_review['last_review'])
print(airbnb_last_review['last_review'].dtype)
print(airbnb_last_review['last_review'].isna().sum())
```

## Question 1: What are the dates of the earliest and most recent reviews? Store these values as two separate variables with your preferred names.
To find the earliest and most recent reviews, I simply found the min and max values in the 'last_review' column of `airbnb_last_review`. The earliest review appears to be from January 1, 2019, and the most recent review is from July 9, 2019. 

```python
earliest_review = airbnb_last_review['last_review'].min()
print(earliest_review)
most_recent_review = airbnb_last_review['last_review'].max()
print(most_recent_review)
```

## Question 2: How many of the listings are private rooms? Save this into any variable.
It looks like 11,356 of the listings are for private rooms. 

```python
num_private_rooms = airbnb_room_type['room_type'].value_counts()['Private Room']
print(num_private_rooms)
```

## Question 3: What is the average listing price? Round to the nearest two decimal places and save into a variable.
Average price per night is $141.78 USD. 

```python
avg_price = round(airbnb_price['price'].mean(), 2)
print(avg_price)
```

## Combine the new variables into one DataFrame called review_dates with four columns in the following order: first_reviewed, last_reviewed, nb_private_rooms, and avg_price. The DataFrame should only contain one row of values.
I first created a dictionary with my desired column name as the key and its variable as the value, then converted that to a DataFrame. 

```python
dict_of_values = {'first_reviewed': [earliest_review], 'last_reviewed': [most_recent_review],\
                  'nb_private_rooms': [num_private_rooms], 'avg_price': [avg_price]}
review_dates = pd.DataFrame(dict_of_values)
print(review_dates)
```
